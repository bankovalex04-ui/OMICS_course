# Project 2 — RNA-seq and CAGE analysis

## Goal

The goal of this project was to generate RNA-seq and CAGE coverage tracks, compare samples in IGV, and identify an example transcriptionally active locus.

Samples: MoPh7, MoPh11, MoPh14, MoPh15.

## Workflow summary

The workflow included:

1. Downloading RNA-seq and CAGE FASTQ files.
2. Building a STAR genome index.
3. Aligning RNA-seq and CAGE reads with STAR.
4. Generating bigWig tracks.
5. Generating an additional BWA-based RNA-seq track for MoPh7 as a comparison.
6. Visualizing tracks in IGV.

```python
# STAR index generation used low-memory parameters
STAR \
  --runThreadN 1 \
  --runMode genomeGenerate \
  --genomeDir day1_HiC_practice/data/reference/star_index \
  --genomeFastaFiles day1_HiC_practice/data/reference/T2T_human.fna \
  --genomeSAindexNbases 10 \
  --genomeSAsparseD 16 \
  --limitGenomeGenerateRAM 8500000000
```

## Selected example locus

## Interpretation

CAGE signal indicates promoter/TSS-associated activity, while RNA-seq reflects transcript-level coverage.  
The MoPh7 STAR and BWA RNA-seq tracks differ, which illustrates why STAR is more appropriate for RNA-seq: it accounts for splicing.

## Important limitation

The tracks were generated against a T2T reference, while IGV visualization was performed using hg38. Therefore, gene labels in IGV should be treated as orientation only, not as strict gene-level annotation.
