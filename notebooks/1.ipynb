# День 1 — Hi-C analysis

## Задача
обработать Hi-C данные и построить контактные карты для четырёх образцов (MoPh7, MoPh11, MoPh14, MoPh15). Образец MoPh7 использован в качестве контроля для визуального сравнения.

## Краткий план работы

Обработка парных ридов, построение .hic файлов, визуализация карт контактов в Juicebox, сравнение полногеномных и хромосомных паттернов контактов.

1. Processing paired-end Hi-C reads.
2. Building `.hic` files.
3. Opening contact maps in Juicebox.
4. Comparing whole-genome and chromosome-level contact patterns.

## Результат
Полногеномные карты демонстрируют ожидаемый сильный диагональный сигнал во всех образцах. Крупных внедиагональных межхромосомных блоков не обнаружено.Отсутствие очевидных крупных межхромосомных перестроек в исследованных образцах по сравнению с контролем.

## Ограничения
карты имеют относительно низкое разрешение, поэтому тонкие структуры (TADы, петли, мелкие перестройки) не могут быть надёжно интерпретированы без более глубокого покрытия и дополнительного статистического анализа.
